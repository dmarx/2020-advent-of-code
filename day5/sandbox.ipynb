```python
def manhattan2binary(in_string):
    binary_str = ''
    for char in in_string.lower():
        outv = '0'
        if char in ('b','r'):
            outv = '1'
        #yield outv
        binary_str += outv
    return int(binary_str, 2)

def pass2seatid(boardingpass, return_index=False):
    row_code, col_code = boardingpass[:7], boardingpass[7:]
    row, col = [manhattan2binary(v) for v in (row_code, col_code)]
    outv = row*8 + col
    if return_index:
        outv = (outv, row, col)
    return outv
```

```python
[manhattan2binary(v) for v in ("BFFFBBF","RRR")], pass2seatid("BFFFBBFRRR")
```

```python
maxv = 0
for rec in records:
    v = pass2seatid(rec)
    maxv = max(v, maxv)
print(maxv)
```
